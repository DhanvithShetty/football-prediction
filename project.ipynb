```python
### We now do rolling averages which us previous results and calculate what result might be for the next week to imporve accuracy ### 
```

```python
def rolling_averages(group,cols,new_cols):
    group=group.sort_values("Date")
    rolling_stats=group[cols].rolling(3,closed='left').mean() #3 weeks,this is to find rolling average till current week(excluding current week(closed=left does this))
    group[new_cols]=rolling_stats
    group=group.dropna(subset=new_cols)  #for removing missing values 
    return group
```

```python
cols = ["GF","GA","Sh","SoT","Dist","FK","PK","PKatt"]
new_cols = [f"{c}_rolling" for c in cols]                #just to make a newcols with cols+rolling
```

```python
new_cols
```

```python
rolling_averages(group,cols,new_cols)
```

```python
matches_rolling = matches.groupby("Team").apply(lambda x: rolling_averages(x,cols,new_cols)) # here now for the entire matches dataset we run the function of rolling averages 
```

```python
def make_predictions(data,predictors):
    train = data[data["Date"]<'2021-08-12']
    test=data[data["Date"]>'2021-08-12']
    rf.fit(train[predictors],train["target"]) 
    preds=rf.predict(test[predictors])   
    combined=pd.DataFrame(dict(actual=test["target"],prediction=preds),index=test.index)
    precision=precision_score(test["target"],preds)
    return combined,precision
```

```python
class MissingDict(dict):
    __missing__=lambda self, key: key
    
map_values = {
    "Brighton and Hove Albion": "Brighton",
    "Manchester United":"Manchester Utd",
    "Newcastle United":"Newcastle Utd",
    "West Ham United":"West Ham",
    "Wolverhampton Wanderers": "Wolves"
}
mapping = MissingDict(**map_values)  ##this is used to make the names proper like if it is arsenal then it will be usual arsenal itself because of lambda self, key: key, but if it is manchester united it will become manchester utd because of mapping 
```
